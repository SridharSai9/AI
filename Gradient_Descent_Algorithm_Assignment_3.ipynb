```python
import numpy as n
import math as m
x=n.random.uniform(-10,10)
y=n.random.uniform(-10,10)
eta=0.01
max=100
i=0
while i<max:
  x_slope=6*x 
  y_slope=-5*m.exp(-y)  
  x_step_length=-eta*x_slope  
  y_step_length=-eta*y_slope
  x=x+x_step_length
  y=y+y_step_length
  i=i+1
min_value=3*m.pow(x,2)+5*m.exp(-y)+10
print("Minimum value of objective function at x =",round(x),"and y =",round(y),"is",round(min_value))
```
